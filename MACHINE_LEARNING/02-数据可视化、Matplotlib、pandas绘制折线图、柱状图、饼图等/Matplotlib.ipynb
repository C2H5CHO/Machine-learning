# 1、`lines` --绘制线条

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5])
y = np.array([3, 6, 9, 12, 15])

# 绘制水平线、垂直线
plt.axhline(
    y=6,
    ls=':', # 虚线
    c='b'
)
plt.axvline(
    x=4,
    ls='-', # 实线
    c='r'
)

# 一次绘制多个线条，并设置每个线条起始终止范围
plt.vlines(
    [2, 3, 3.5],  # 三个线条的x坐标
    [10, 20, 30],  # 三个线条起始点y坐标
    [25, 35, 45]  # 三个线条结束点y坐标
)

plt.plot(x, y)
plt.show()
```

# 2、`sin_cos` --绘制正弦、余弦函数

```python
import numpy as np
import matplotlib.pyplot as plt
import math

x = np.arange(0, 2*np.pi, 0.1)
print('x shape：', x.shape)

y1 = np.sin(x) # 正弦曲线
y2 = np.cos(x) # 余弦曲线
print('y1 shape：', y1.shape)
print('y2 shape：', y2.shape)

# label：图例名称
# linewidth：线宽
# linestyle：线型
plt.plot(x, y1, label='sin', linewidth=2.0)
plt.plot(x, y2, label='cos', linewidth=4.0, linestyle='--')

# 设置坐标轴名称
plt.xlabel('x')
plt.ylabel('y')

# 设置图标题
plt.title('sin and cos', fontsize=20)

# 设置坐标轴范围
plt.xlim(0, 2*math.pi)
plt.ylim(-1, 2)

plt.legend() # 显示图例
plt.show()
```

# 3、`xyticks` --设置刻度显示的文字

```python
import numpy as np
import matplotlib.pyplot as plt
import math

x = np.arange(-5, 5, 0.1)
y = x**2

plt.plot(x, y, linewidth=2.0, color='r', alpha=0.5, label='y=x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-10, 10)
plt.ylim(-1, 30)
plt.title('$y=x^2$', fontsize=20)

# 设置刻度
x_ticks = np.arange(-10, 11, 2)
x_ticks = x_ticks.astype(int) # 转换为整数类型
y_ticks = np.arange(-1, 31, 5)
plt.xticks(x_ticks)
plt.yticks(y_ticks)

# 添加标记点--实质：绘制散点图
x_arr = np.arange(-6, 6+1, 2)
y_arr = x_arr**2 # y坐标值
plt.scatter(
    x_arr, # x坐标值
    y_arr, # y坐标值
    marker='o', # 标记点形状
    s=100, # 标记点大小
    c='b', # 标记点颜色
    alpha=0.5 # 标记点透明度
)

# loc参数设置标签位置： best, upper right/upper left, lower left/lower right, right, center left/center right, lower center/upper center, center
plt.legend(loc='upper left')
plt.show()
```

# 4、`gca` --设置坐标轴格式

```python
import matplotlib.pyplot as plt

ax = plt.gca() # 获取当前坐标轴字典
axis_b = ax.spines['bottom'] # 获取下轴
axis_b.set_position(('data', 0)) # 0表示x轴的起始位置
axis_l = ax.spines['left'] # 获取左轴
axis_l.set_position(('data', 0)) # 0表示y轴的起始位置

ax.spines['top'].set_color('none') # 设置上轴无色
ax.spines['right'].set_color('none') # 设置右轴无色

plt.show()
```

# 5、`figure` --创建多个图形窗体

```python
import matplotlib.pyplot as plt

# 第一个
plt.figure(
    'FigA', # 图形名称
    facecolor='w', # 背景色
)
plt.grid(linestyle='--') # 显示网格

# 第二个
plt.figure(
    'FigB', # 图形名称
    facecolor='gray', # 背景色
    figsize=(25, 15), # 图形大小
)
plt.xlabel('Date', fontsize=14) # 设置x轴名称
plt.ylabel('Price', fontsize=14) # 设置y轴名称
plt.grid(linestyle='-.') # 显示网格
plt.tight_layout() # 自动调整子图间距

plt.show() # 显示图形
```

# 6、`subplot` --子图

```python
import matplotlib.pyplot as plt

plt.figure('SubplotDemo', facecolor='lightgray', figsize=(25, 15)) # 图形名称和背景色
for i in range(9):
    plt.subplot(3, 3, i+1) # 3行3列第i+1个子图

    plt.text(
        0.5, 0.5, # 文本位置
        i+1, # 文本内容
        ha='center', # 水平对齐方式
        va='center', # 垂直对齐方式
        fontsize=25, # 字体大小
        alpha=0.5, # 透明度
        color='red' # 字体颜色
    )

    plt.xticks([]) # 隐藏x轴刻度
    plt.yticks([]) # 隐藏y轴刻度

plt.show() # 显示图形
```
