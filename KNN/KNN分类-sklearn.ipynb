```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 样本特征
data_x = [
    [1.3, 6],
    [3.5, 5],
    [4.2, 2],
    [5, 3.3],
    [2, 9],
    [5, 7.5],
    [7.2, 4],
    [8.1, 8],
    [9, 2.5]
]

# 样本标签
data_y = [0, 0 ,0 ,0, 1, 1, 1, 1, 1]

# 训练集
x_train = np.array(data_x)
y_train = np.array(data_y)

print(x_train)
print(y_train)

plt.scatter(x_train[y_train == 0, 0], x_train[y_train == 0, 1], color='r', marker='o')
plt.scatter(x_train[y_train == 1, 0], x_train[y_train == 1, 1], color='g', marker='x')
plt.show()
```

```python
# 新样本
data_new = np.array([4, 5])

plt.scatter(x_train[y_train == 0, 0], x_train[y_train == 0, 1], color='r', marker='o')
plt.scatter(x_train[y_train == 1, 0], x_train[y_train == 1, 1], color='g', marker='x')
plt.scatter(data_new[0], data_new[1], color='b', marker='^')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_clf  = KNeighborsClassifier(n_neighbors=5)
knn_clf.fit(x_train, y_train)
knn_clf.predict([data_new])
```
